```python
#Masala 0006
yil = int(input())
kabisa = False
nol = '0'*(4 - len(str(yil)))
if yil%4==0:
    kabisa=True
if yil%100==0 and yil%400 !=0:
    kabisa=False
if kabisa:
    print(f'12/09/{nol}{yil}')
else:
    print(f'13/09/{nol}{yil}')    
```

```python
#Masala 0010
N = int(input())
a,b,c=map(int,input().split(' '))
if a+b+c>=N:
    print('Yes')
else:
    print('No')
```

```python
#Masala 0011
n = int(input())
lst =list(map(int, input().split(' ')))
m = max(lst)
lst = [m - i for i in lst]
if lst.count(0) >=2:
    print(m)
else:
    lst.remove(0)
    print(m-min(lst))
```
