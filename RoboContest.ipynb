```python
#Masala 0005

z = int(input())
if z==0:
    print(-1)
else:    
    s = 0
    num = [i for i in range((-1)*abs(z),abs(z)+1) if i!=0 and z%i==0]
    for i in num:
        for j in num:
            if i*j==z and j>=i:
                s+=1
    print(s)  
```

```python
#Masala 0006
yil = int(input())
kabisa = False
nol = '0'*(4 - len(str(yil)))
if yil%4==0:
    kabisa=True
if yil%100==0 and yil%400 !=0:
    kabisa=False
if kabisa:
    print(f'12/09/{nol}{yil}')
else:
    print(f'13/09/{nol}{yil}')    
```
