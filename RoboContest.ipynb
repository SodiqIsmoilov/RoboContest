```python
#Masala 0006
yil = int(input())
kabisa = False
nol = '0'*(4 - len(str(yil)))
if yil%4==0:
    kabisa=True
if yil%100==0 and yil%400 !=0:
    kabisa=False
if kabisa:
    print(f'12/09/{nol}{yil}')
else:
    print(f'13/09/{nol}{yil}')    
```
